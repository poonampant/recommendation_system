```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings=pd.read_csv("ratings.csv")
print(ratings)
movies=pd.read_csv("movies.csv")
print(movies.head())
ratings=pd.merge(movies,ratings)
print(ratings.head())
df=movies
```

```python
"""select features"""
```

```python
features=['genres']
```

```python
for feature in features:
    df[feature]=df[feature].fillna('')

def combine_feature(row):
    return row['genres']

df["combine_feature"]=df.apply(combine_feature,axis=1)
print(df["combine_feature"])
```

```python
cv=CountVectorizer()
count_matrix=cv.fit_transform(df["combine_feature"])
```

```python
cosine_sim=cosine_similarity(count_matrix)
print(cosine_sim)
```

```python
movie_index=(ratings[ratings["userId"]==1 ]["movieId"].values[1])
print(movie_index)
```

```python
movie_user_like=(df[df["movieId"]== movie_index]["title"].values[0])
print(movie_user_like)
```

```python
"""get index of movie Avatar and then print all the movie which are corellated to that movie in decreasing order"""
```

```python
similar_movies=list(enumerate(cosine_sim[movie_index]))
print(similar_movies)
sorted_similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)

def get_title_from_index(index):
    return (df[df["movieId"]== index]["title"].values[0])
```

```python
i=0
for movie in sorted_similar_movies:
    print(get_title_from_index(movie[0]))
    i=i+1
    if i>8:
        break
```
